```python
import pandas as pd
```

```python
df = pd.read_csv('../datasets/HI-Small_Trans_2banks+3others_SEK.csv', parse_dates=['Timestamp'])
```

```python
accounts = set(df['Account']).union(set(df['Account.1']))
account_id_map = {account: i+1 for i, account in enumerate(sorted(accounts))}
df['from_account_id'] = df.apply(lambda row: account_id_map.get(row['Account'], 0), axis=1)
df['to_account_id'] = df.apply(lambda row: account_id_map.get(row['Account.1'], 0), axis=1)
```

```python
banks = set(df['From Bank']).union(set(df['To Bank']))
bank_id_map = {bank: i+1 for i, bank in enumerate(sorted(banks))}
df['from_bank_id'] = df.apply(lambda row: bank_id_map.get(row['From Bank'], 0), axis=1)
df['to_bank_id'] = df.apply(lambda row: bank_id_map.get(row['To Bank'], 0), axis=1)
```

```python
df.rename(columns={
    'Timestamp': 'timestamp', 
    'From Bank': 'from_bank', 
    'To Bank': 'to_bank', 
    'Account': 'from_account', 
    'Account.1': 'to_account', 
    'Amount Received': 'amount',
    'Is Laundering': 'is_laundering'}, inplace=True)
df.drop(columns=['Amount Paid', 'Receiving Currency', 'Payment Currency', 'Payment Format'], inplace=True)
df.sort_values(by=['timestamp'], inplace=True)
df = df[[
    'timestamp', 
    'from_account', 
    'from_account_id', 
    'from_bank', 
    'from_bank_id', 
    'to_account', 
    'to_account_id',
    'to_bank', 
    'to_bank_id',
    'amount', 
    'is_laundering']]
```

```python
df.sort_values(by=['timestamp'], inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df.to_parquet('../datasets/transactions_small.parquet', index=False)
df
```
