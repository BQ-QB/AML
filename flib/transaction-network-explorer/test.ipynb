```python
import pandas as pd
```

```python
df = pd.read_csv('../AMLsim/outputs/10K_accts/tx_log.csv')
df1 = df[['nameOrig', 'step', 'amount', 'newbalanceOrig']]
df1 = df1.rename(columns={'nameOrig': 'name'})
df1['amount'] = -df1['amount']
df2 = df[['nameDest', 'step', 'amount']]
df2 = df2.rename(columns={'nameDest': 'name'})

df = pd.concat([df1, df2]).reset_index(drop=True)
df = df.sort_values(by=['step'])
df['balance'] = df.groupby('name')['amount'].cumsum()
a1 = df[df['name']==1]

a1.plot(x='step', y='balance', grid=True, figsize=(10, 5))
        
```

```python
df = pd.read_csv('../AMLsim/outputs/10K_accts/tx_log.csv')
df1 = df[['nameOrig', 'step', 'newbalanceOrig']]
df1 = df1.rename(columns={'nameOrig': 'name', 'newbalanceOrig': 'balance'})
df2 = df[['nameDest', 'step', 'newbalanceDest']]
df2 = df2.rename(columns={'nameDest': 'name', 'balance': 'newbalanceDest'})

df = pd.concat([df1, df2])
a1 = df[df['name']==1]
a1.plot(x='step', y='balance', grid=True, figsize=(10, 5))
```
